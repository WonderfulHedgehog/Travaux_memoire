## Tertullien

### Pallio

```python
xmlstr = ET.parse('Pall2.xml')
root = xmlstr.getroot()

for deep2 in root[1]:
    print(deep2.attrib, deep2.tag)

new_root = root[1][0]

for deep3 in new_root:
    print(deep3.attrib, deep3.tag)

for deep4 in new_root[2]:
    print(deep4.attrib, deep4.tag)

for i in range (6):
    for deep5 in new_root[2][i]:
        print(deep5.attrib, deep5.tag)

# for i in range (6):
#     print("Chapitre " + str(i+1))
#     print("\n")
#     for deep6 in new_root[2][i][0]:
#         print(deep6.attrib, deep6.tag)
```

```python
TXT_metaphore=""
TXT_epiphoneme=""
TXT_comparaison=""
TXT_personnification=""
TXT_enargeia=""
TXT_ab_absurdo=""
TXT_ab_origine=""
TXT_calembour=""
TXT_dicton=""
TXT_dialogisme=""
TXT_prosopopeia=""
for i in range (6):
    print("Chapitre " + str(i+1))
    print("\n")
    #Je précise les chapitres dans les sorties txt
    TXT_metaphore += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_epiphoneme += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_comparaison += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_personnification += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_enargeia += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_ab_absurdo += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_ab_origine += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_calembour += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_dicton += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_dialogisme += "\n" + "Chapitre " + str(i+1) + "\n"
    TXT_prosopopeia += "\n" + "Chapitre " + str(i+1) + "\n"
    
    for deep6 in new_root[2][i][0]:
        if deep6.attrib.get("ana") != None:
            print("Tag de la balise : " + deep6.attrib.get("ana"))
            print("\n")
            print(deep6.text)
            if deep6.attrib.get("ana") == "metaphore" :
                TXT_metaphore += deep6.text
            if deep6.attrib.get("ana") == "epiphoneme" :
                TXT_epiphoneme += deep6.text
            if deep6.attrib.get("ana") == "comparaison" :
                TXT_comparaison += deep6.text
            if deep6.attrib.get("ana") == "personnification" :
                TXT_personnification += deep6.text
            if deep6.attrib.get("ana") == "enargeia" :
                TXT_enargeia += deep6.text
            if deep6.attrib.get("ana") == "ab_absurdo" :
                TXT_ab_absurdo += deep6.text
            if deep6.attrib.get("ana") == "ab_origine" :
                TXT_ab_origine += deep6.text
            if deep6.attrib.get("ana") == "calembour" :
                TXT_calembour += deep6.text
            if deep6.attrib.get("ana") == "dialogisme" :
                TXT_dialogisme += deep6.text
            if deep6.attrib.get("ana") == "prosopopeia" :
                TXT_prosopopeia += deep6.text
        else:
            # print("Texte de la balise :" + deep6.attrib.get("type"))
            # print("\n")
            # print(deep6.text)
            if deep6.attrib.get("type") == "dicton?" :
                TXT_dicton += deep6.text
```

```python
#Ecriture des fichiers TXT
file = open("Pallio/Tertullien_metaphore.txt","w") 
file.writelines(TXT_metaphore)
file.close()

file = open("Pallio/Tertullien_epiphoneme.txt","w") 
file.writelines(TXT_epiphoneme)
file.close()

file = open("Pallio/Tertullien_comparaison.txt","w") 
file.writelines(TXT_comparaison)
file.close()

file = open("Pallio/Tertullien_personnification.txt","w") 
file.writelines(TXT_personnification)
file.close()

file = open("Pallio/Tertullien_enargeia.txt","w") 
file.writelines(TXT_enargeia)
file.close()

file = open("Pallio/Tertullien_ab_absurdo.txt","w") 
file.writelines(TXT_ab_absurdo)
file.close()

file = open("Pallio/Tertullien_ab_origine.txt","w") 
file.writelines(TXT_ab_origine)
file.close()

file = open("Pallio/Tertullien_calembour.txt","w") 
file.writelines(TXT_calembour)
file.close()

file = open("Pallio/Tertullien_dialogisme.txt","w") 
file.writelines(TXT_dialogisme)
file.close()

file = open("Pallio/Tertullien_prosopopeia.txt","w") 
file.writelines(TXT_prosopopeia)
file.close()

file = open("Pallio/Tertullien_dicton.txt","w") 
file.writelines(TXT_dicton)
file.close()
```
